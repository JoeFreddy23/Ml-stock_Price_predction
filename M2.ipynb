```python
import pandas as pd 
import numpy as n
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d = pd.read_csv('A.csv')
d.head()
```

```python
d.columns
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(d['open'],color='red')

plt.subplot(1,2,2)
plt.hist(d['high'],color='blue')
```

```python
d.hist(figsize=(15,55),layout=(15,4), color='red')
plt.show()
```

```python
d['low'].hist(figsize=(10,5),color='red')
```

```python
plt.bar(d['close'],d['volume'], color='red') # scatter, plot, triplot, stackplot
```

```python
plt.boxplot(d['open']) 
```

```python
d['high'].plot(kind='density')
```

```python
sns.displot(d['low'], color='red') 
# barplot, boxenplot, boxplot, countplot, displot, distplot, ecdfplot, histplot, kdeplot, pointplot, violinplot, stripplot
```

```python
sns.residplot(d['close'],d['volume'], color='red') # residplot, scatterplot 
```

```python
sns.pairplot(d)
```

```python
fig, ax = plt.subplots(figsize=(20,15))
sns.heatmap(d.corr(),annot = True, fmt='0.2%',cmap = 'autumn',ax=ax)
```
